```python
target = loans_df['loan_status_float'].as_matrix()
target
```

```python

input_data = loans_df_new.as_matrix()
loans_df_new
```

```python

from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(input_data, target, test_size=0.33, random_state=42)
input_data.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10)
rf.fit(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

```python
rf.feature_importances_
```

```python
from sklearn.cross_validation import cross_val_score
from sklearn.ensemble import ExtraTreesClassifier
clf = ExtraTreesClassifier(n_estimators=10, max_depth=None,min_samples_split=2, random_state=0)
scores = cross_val_score(clf, input_data, target)
scores.mean()
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(C=1e5)
logreg.fit(X_train, y_train)
logreg.score(X_test, y_test)
```

```python
from sklearn.linear_model import SGDClassifier
sgd = SGDClassifier()
sgd.fit(X_train, y_train)
sgd.score(X_test, y_test)
```
