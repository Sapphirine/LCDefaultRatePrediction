```python
import pandas as pd
import numpy as np

loans_df = pd.read_csv('/home/niteshch/Downloads/loan.csv')
```

```python
loans_df['annual_inc_val'] = np.where(loans_df['application_type']== 'INDIVIDUAL', loans_df['annual_inc'], loans_df['annual_inc_joint'])
loans_df['dti_val'] = np.where(loans_df['application_type']== 'INDIVIDUAL', loans_df['dti'], loans_df['dti_joint'])
loans_df.count()

```

```python
loans_df.replace('n/a', np.nan,inplace=True)
#loans_df.drop(['total_bal_il','il_util','open_rv_12m','open_rv_24m','max_bal_bc','all_util','total_rev_hi_lim','inq_fi','total_cu_tl','inq_last_12m','acc_now_delinq','tot_coll_amt','tot_cur_bal'], axis=1, inplace=True)
loans_df.dropna(axis=1, how='all', thresh=100000, inplace=True)
loans_df.count()
```

```python
loans_df.drop(['emp_title','url','desc','title','zip_code','dti','annual_inc'], axis=1, inplace=True)
```

```python
loans_df.fillna(np.nan, axis=1, inplace=True)
loans_df.count()
```

```python
loans_df.to_csv('/home/niteshch/Downloads/loan_cleaned_1.csv', header=True, index=False)
```

```python
loans_df.columns.to_series().groupby(loans_df.dtypes).groups
```

```python
loans_df['term_float'] = np.where(loans_df['term'] == ' 36 months', np.float64(36), np.float64(60))
loans_df[['term','term_float']]
```

```python
def transform_grade(row):
    if row['grade'] == 'A':
        return np.float64(1)
    elif row['grade'] =='B':
        return np.float64(2)
    elif row['grade'] =='C':
        return np.float64(3)
    elif row['grade'] =='D':
        return np.float64(4)
    elif row['grade'] =='E':
        return np.float64(5)
    elif row['grade'] =='F':
        return np.float64(6)
    elif row['grade'] =='G':
        return np.float64(7)

loans_df['grade_float'] = loans_df.apply(transform_grade, axis=1)
loans_df[['grade', 'grade_float']]
```

```python
loans_df['loan_status_float'] = np.float64(0)
loans_df['loan_status'].unique()
```

```python
loans_df['loan_status_float'] = [np.float64(1) if x == 'Fully Paid' or x == 'Current' 
                                 else np.float64(0) for x in loans_df['loan_status']]

loans_df[['loan_status', 'loan_status_float']]
```

```python
loans_df_new = loans_df[loans_df.loan_status != 'Does not meet the credit policy. Status:Fully Paid']
loans_df_new = loans_df_new[loans_df.loan_status != 'Does not meet the credit policy. Status:Charged Off']
loans_df_new = loans_df_new[loans_df.loan_status != 'Issued']
loans_df_new['loan_status'].unique()
loans_df_new[['loan_status', 'loan_status_float']]
```

```python
import time
def convert_date(row):
    return time.mktime(time.strptime(row['issue_d'], '%b-%Y'))

loans_df_new['issue_date_epoch'] = loans_df_new.apply(convert_date, axis=1)
```

```python
loans_df_new['application_type_float'] = np.where(loans_df_new['application_type'] == 'INDIVIDUAL', 
                                                  np.float64(1), np.float64(2))
```

```python
loans_df_new.drop(['term', 'grade', 'sub_grade', 'emp_length', 'home_ownership',
        'verification_status', 'issue_d', 'loan_status', 'pymnt_plan',
        'purpose', 'addr_state', 'earliest_cr_line', 'initial_list_status',
        'last_pymnt_d', 'next_pymnt_d', 'last_credit_pull_d',
        'application_type'], axis=1, inplace=True)
```

```python
loans_df_new.drop(['id', 'member_id'], axis=1, inplace=True)
```

```python
loans_df_new.columns.to_series().groupby(loans_df.dtypes).groups
```

```python
loans_df_new.to_csv('/home/niteshch/Downloads/loan_cleaned_2.csv', header=True, index=False)
```
